```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/canada_per_capita_income.csv")
```

```python
data.head(10)
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.scatter(data.year, data["per capita income (US$)"], color = "red", marker = "+")
```

```python
df = data.drop('per capita income (US$)',axis='columns')
df.head(5)
```

```python
income = data["per capita income (US$)"]
income.head(5)
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(df,income)
```

```python
reg.predict([[2020]])
```
