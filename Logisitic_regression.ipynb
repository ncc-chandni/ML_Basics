```python
df.columns
```

```python
subdf = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
subdf.head()
```

Tackle salary dummy variable

Salary has all text data. It needs to be converted to numbers and I will use dummy variable for that.

```python
salary_dummies = pd.get_dummies(subdf.salary, prefix = "salary")
```

```python
df_with_dummies = pd.concat([subdf,salary_dummies], axis = 'columns')
df_with_dummies.head()
```

Now I can remove salary column which is text data. It is already replaced by dummy variables so I can safely remove it



```python
df_with_dummies.drop('salary', axis = 'columns', inplace = True)
df_with_dummies.head()
```

```python
X = df_with_dummies
X.head()
```

```python
y = df.left
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

Accuracy of the model

```python
model.score(X_test, y_test)
```
