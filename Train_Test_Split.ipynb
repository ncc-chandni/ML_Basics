Looking at above two scatter plots, using linear regression model makes sense as it can be seen clearly a linear relationship between our dependant (i.e. Sell Price) and independant variables (i.e. car age and car mileage)

The approach I am going to use here is to split available data in two sets

Training: I will train my model on this dataset
Testing: I will use this subset to make actual predictions using trained model
The reason I don't use same training set for testing is because my model has seen those samples before, using same samples for making predictions might give us wrong impression about accuracy of my model. It is like you ask same questions in exam paper as you tought the students in the class.

```python
X = df[["Mileage","Age(yrs)"]]
```

```python
y = df["Sell Price($)"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
y_train
```

```python
y_test
```

Let's run Linear Regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

Random State Argument

```python
#random_state produces same sample everytime
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
X_test
```
